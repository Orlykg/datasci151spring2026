**Q2** Conversion of interest rates

<font size = "3">

**NOTE:** For this problem, interest rates are always given in decimal form. An interest rate of 25% corresponds to $r = 0.25$, an interest rate of 5% corresponds to $r = 0.05$, etc.

Suppose you have a nominal annual interest rate $r_1$ that is compounded $n_1$ times per year. Then the value of $1 after one year is given by:

$$ \left(1 + \frac{r_1}{n_1} \right)^{n_1}$$

If you want to calculate the equivalent interest rate $r_2$ with a *different* compounding frequency $n_2$, you would equate the value of a dollar after one year in both cases:

$$ \left(1 + \frac{r_2}{n_2} \right)^{n_2} = \left(1 + \frac{r_1}{n_1} \right)^{n_1}$$

and then solve for $r_2$:

\begin{align}
 r_2 = n_2\cdot \left[ \left(1 + \frac{r_1}{n_1} \right)^{\frac{n_1}{n_2}} - 1 \right]
 \end{align}

- Using the equation above as a guide, write a function which calculates the value of $r_2$ based on the values of $r_1$, $n_1$, and $n_2$ 

- Then evaluate the function to determine what interest rate we would need to convert a 10% interest rate ($r_1 = 0.1$) compounded daily ($n_1 = 365$) into one that is compounded monthly ($n_2 = 12$)?

**NOTE: Read the note at the top of this question.**

```python
# write your code below

def investment_value(r1, r2, n1, n2): 
    r2 = n2 * ((1 + r1/n1)**(n1/n2) - 1)
    return r2

print(investment_value(r1 = 0.1, n1 = 365, n2 = 12, r2 = 0))

#check 
r2 = 12 * ((1 + 0.1/365)**(365/12) - 1)
print(r2)
```

**Q4:** A continous random variable $X$ is said to have an **exponential distribution with parameter $\lambda > 0$**, if its probability density function (pdf) $f_X$ has the form:

$$ f_X(x) = \begin{cases} \lambda \mathrm{e}^{-\lambda x}\quad \textrm{when } x > 0\\
0\qquad\quad \textrm{when } x \leq 0\end{cases} $$

- Complete the function below, implementing the probability density function $f_X$ as a Python function.

- Your function should be able to handle any floating point number for the first argument `x`, and any **positive** floating point number for the the second argument `lam`.

- Note: Python has a built-in keyword called `lambda`, so we cannot use that as the name of an input argument (it would cause a SyntaxError)

```python
import numpy as np 

# complete the function below

def pdf_exp(x, lam):
    fx1 = lam* np.exp(-lam*x)
    if lam > 0: 
        return fx1
    else:
        return 0

#check
print(pdf_exp(1, 2))
print(pdf_exp(-1, 2))
print(pdf_exp(1, -1))
```
