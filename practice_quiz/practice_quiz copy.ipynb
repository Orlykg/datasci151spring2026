• To get full points, plot the scatter plots on a single column using the subplots syntax (see Lecture 6)

```python
n = 500

# Generate random variables
vec_a = np.random.poisson(lam=3, size=n)
vec_b = np.random.exponential(scale=2, size=n)
vec_c = np.random.binomial(n=10, p=0.4, size=n)

# Create subplots (single column)
fig, axes = plt.subplots(2, 1, figsize=(6, 10))

# First scatter plot (vec_b vs vec_a)
axes[0].scatter(vec_a, vec_b)
axes[0].set_xlabel("vec_a (Poisson)")
axes[0].set_ylabel("vec_b (Exponential)")
axes[0].set_title("Scatter Plot: vec_b vs vec_a")

# Second scatter plot (vec_c vs vec_a)
axes[1].scatter(vec_a, vec_c)
axes[1].set_xlabel("vec_a (Poisson)")
axes[1].set_ylabel("vec_c (Binomial)")
axes[1].set_title("Scatter Plot: vec_c vs vec_a")

plt.tight_layout()
plt.show()
```

• Let n = 1000
• Generate vec_x from a Normal with (loc = 2, scale = 1, size = n)
• Generate vec_z from a chi square with (df = 2, size = n)
• Generate vex_y from a uniform with (low = 5, high = 10, size = n)
• Plot two scatter plots: ( vec_x vs vec_y) and (vec_x vs vec_z)
• Label the axes and title
• Each graph should have a different title

```python
import numpy as np
import matplotlib.pyplot as plt

# Set n
n = 1000

# Generate random variables
vec_x = np.random.normal(loc=2, scale=1, size=n)
vec_z = np.random.chisquare(df=2, size=n)
vec_y = np.random.uniform(low=5, high=10, size=n)

# First scatter plot: vec_x vs vec_y
plt.scatter(vec_x, vec_y)
plt.xlabel("vec_x (Normal)")
plt.ylabel("vec_y (Uniform)")
plt.title("Scatter Plot: Normal vs Uniform")
plt.show()

# Second scatter plot: vec_x vs vec_z
plt.scatter(vec_x, vec_z)
plt.xlabel("vec_x (Normal)")
plt.ylabel("vec_z (Chi-Square)")
plt.title("Scatter Plot: Normal vs Chi-Square")
plt.show()
```
